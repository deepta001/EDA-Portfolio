```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('train.csv')

df.head()
```

```python
sns.countplot(x='Survived', data=df)
plt.show()
```

More passengers died than survived on the Titanic.

```python
sns.barplot(x='Sex', y='Survived', data=df)
plt.show()
```

Female passengers had a significantly higher survival rate than male passengers.

```python
sns.barplot(x='Pclass', y='Survived', data=df)
plt.title('Survival Rate by Passenger Class')
plt.show()
```

First-class passengers had the highest survival rates, while third-class passengers had the lowest.

```python
sns.histplot(df['Age'], bins=30, kde=True)
plt.title('Age Distribution')
plt.show()
```

Most passengers were between 20 and 40 years old.

```python
sns.boxplot(x=df['Fare'])
plt.title('Fare Distribution')
plt.show()
```

Most passengers paid relatively low fares, with several extreme high-fare outliers.

```python
sns.boxplot(x='Survived', y='Fare', data=df)
plt.title('Fare vs Survival')
plt.show()
```

Passengers who paid higher fares generally had higher survival rates.

```python
sns.barplot(x='Embarked', y='Survived', data=df)
plt.title('Survival Rate by Embarkation Port')
plt.show()
```

Survival rates varied by embarkation port, likely reflecting passenger class distributions.

```python
df['FamilySize'] = df['SibSp'] + df['Parch'] + 1

sns.barplot(x='FamilySize', y='Survived', data=df)
plt.title('Survival Rate by Family Size')
plt.show()
```

Small families often had better survival rates than solo travelers or very large families.

```python
sns.scatterplot(
    x='Age',
    y='Fare',
    hue='Survived',
    data=df
)

plt.title('Age vs Fare by Survival')
plt.show()
```

Key Insights

 Higher fares are generally associated with higher survival rates. Most passengers paid fares below 50. Children appear to have higher survival rates. Age alone does not strongly predict survival. Passengers paying very high fares were mostly survivors.
